```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ruta = 'https://github.com/juliandariogiraldoocampo/analisis_taltech/raw/refs/heads/main/explorador/top_20_paises.csv'
df = pd.read_csv(ruta)
```

```python
df.head()
```

```python
# Crear una nueva figura para el gráfico con un tamaño personalizado
# (10 de ancho por 6 de alto).
plt.figure(figsize=(10, 6))

# Crear un gráfico de barras utilizando la columna 'Pais' para el eje x
# y 'Promedio Renovables' para el eje y. El color de las barras se establece en
# 'skyblue' para mejorar la visualización.
plt.bar(df['Pais'], df['Promedio Renovables'], color='skyblue')

# Añadir una etiqueta al eje x para indicar que representa los países.
plt.xlabel('País')

# Añadir una etiqueta al eje y para mostrar que representa el promedio de
# producción renovable en porcentaje.
plt.ylabel('Promedio de Producción Renovable (%)')

# Establecer un título para el gráfico que explique el contenido del mismo.
plt.title('Producción Promedio de Energía Renovable por País')

# Rotar las etiquetas del eje x 45 grados para facilitar su lectura.
# El argumento 'ha="right"' alinea las etiquetas hacia la derecha.
plt.xticks(rotation=45, ha='right')

# Ajustar automáticamente el diseño para evitar que las etiquetas se solapen o
# que el gráfico quede desordenado.
plt.tight_layout()

#
#for i, valor in enumerate(df['Promedio Renovables']):
#    plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')

# Mostrar el gráfico en pantalla.
plt.show()
```

Opcional: Crear un gráfico circular para observar la proporción de producción por país:

```python
# Crear una nueva figura para el gráfico circular con un tamaño personalizado
# de 8x8 unidades.
plt.figure(figsize=(8, 8))

# - 'df['Promedio Renovables']': Define los tamaños de los sectores del gráfico.
# - 'labels=df['Pais']': Asigna los nombres de los países como etiquetas
#                        para cada sector.
# - 'autopct='%1.1f%%'': Muestra el porcentaje de cada sector con un decimal.
# - 'startangle=0': Comienza el gráfico desde el ángulo indicado.
# - 'colors=plt.cm.Paired.colors': Utiliza una paleta de colores predefinida
#                                  para diferenciar los sectores.
plt.pie(df['Promedio Renovables'],
        labels=df['Pais'],
        autopct='%1.1f%%',
        startangle=48,
        colors=plt.cm.Paired.colors)

# Establecer un título para el gráfico para contextualizar la información
# representada.
plt.title('Proporción de Producción de Energía Renovable por País')

# Asegurar que el gráfico tenga una forma circular perfecta, en lugar de un óvalo.
plt.axis('equal')

plt.show()
```

```python
from scipy.stats import kurtosis
```

```python

# Curtosis usando pandas
curtosis_pandas = df['Promedio Renovables'].kurt()  # Curtosis excesiva
print(f'Curtosis (pandas): {curtosis_pandas}')

# Curtosis usando scipy
curtosis_scipy = kurtosis(df['Promedio Renovables'], fisher=True, bias=False)  # fisher=True devuelve curtosis excesiva
print(f'Curtosis (scipy): {curtosis_scipy}')
```
